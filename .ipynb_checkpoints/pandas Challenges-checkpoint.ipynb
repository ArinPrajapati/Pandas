###### Diplay summary statistics:
 Use the `.describe()` method to get summary statistics for numerical columns like mean, standard deviation, minimum, maximum, etc.

```python
gendf.describe()
```

##### Check Unique Values:
To find unique values in a specific column, use the `.unique()` method. For example, to find unique values in a column named `ascension_boss `:

```python
gendf['ascension_boss'].unique()
```

#### Value Counts:
To get a count of unique values in a specific column, use the `.value_counts()` method. For example, to count the occurrences of each category in the 'affiliation' column:

```python
gendf['affiliation'].value_counts()
```

```python
# finding a specific value in large data
filtered_df = gendf[gendf['affiliation']=='"The Red Strings"' ]
if not filtered_df.empty:
    index = filtered_df.index[0]
    print("Index of The Red Strings   is:", index)
else:
    print("Value not found in the DataFrame.")
```
